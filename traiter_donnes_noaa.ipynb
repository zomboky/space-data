```python
from netCDF4 import Dataset
import os
```

```python
base = "Donnees Spatiales"

datasets = []

for year in range(2020, 2025):  # 2020 → 2024 inclus
    folder = f"{year}_NOAA_15"
    path_folder = os.path.join(base, folder)

    for filename in os.listdir(path_folder):
        if filename.startswith("poes_n15_") and filename.endswith(".nc"):
            path_file = os.path.join(path_folder, filename)
            ds = Dataset(path_file, "r")
            datasets.append(ds)
            print("Chargé :", path_file)

print(f"\nTotal datasets chargés : {len(datasets)}")
```

```python
# Sauvegarde
datasets.to_csv("./Donnees Spatiales/datasets.csv", index=False)
```
