```python
from sunpy.net import Fido
from sunpy.net import attrs as a
from sunpy.timeseries import TimeSeries
```

```python
trange = a.Time('2020/01/01', '2024/01/01')
dataset = a.cdaweb.Dataset('NOAA15_POES-SEM2_FLUXES-2SEC')
result = Fido.search(trange, dataset)
```

```python
result
```

```python
files = Fido.fetch(result, path="project_data/{file}")
```

```python
# On concatne toutes les annes et on sauvegarde sous forme de pd dataframe
ts = TimeSeries(files, concatenate=True)
df_sunpy = ts.to_dataframe()
```
